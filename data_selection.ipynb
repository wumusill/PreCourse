### basic, loc, iloc selection

```python
df[["CRIM", "AGE"]][:2]  # Column과 index number
```

```python
df.loc[[2, 4, 6], ["CRIM", "AGE"]] # Column과 index name
```

```python
df.iloc[:10, :2]  # Column number와 index number
```
